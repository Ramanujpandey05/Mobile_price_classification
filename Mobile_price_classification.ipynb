# TASK 5 - Create a fully connected neural network (FCN)

```python
import tensorflow as tf
from  tensorflow.keras.models import Sequential #To Link the different layers

from tensorflow.keras.layers import Dense #To create the neuron Layers
```

```python
#Number of Features

num_features = len(data.columns) - 1 #Number of input neurons
num_classes = data["price_range"].nunique() #Number of  output neurons
```

```python
print(num_features,num_classes)
```

```python
#BUILDING THE NEURAL NETWORK M0DEL

model = Sequential(
    [
        Dense(60, activation = "relu", input_shape=(num_features,)),#1st hidden layer with 60 neurons
        Dense(40, activation = "relu"),#2nd hidden layer with 40 neurons
        Dense(num_classes, activation = "softmax") #Output layer with 4 output neurons
    ]
)
    
```

```python
#Compile NN
model.compile(
    optimizer = "adam", #enhanced version of gradient descent "sgd"
    loss = "sparse_categorical_crossentropy",
    metrics = ["accuracy"]
)
```

```python
model.summary()
```

```python
#train
```

```python
model.fit(X,Y,epochs =10,batch_size= 50)
```

```python
model.save("model.keras")
```
